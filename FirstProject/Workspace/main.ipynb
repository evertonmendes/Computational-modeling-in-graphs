```python

    def CreateMatrix(n, p):

        all_combinations=list(itertools.combinations(range(n), 2))
        matrix=np.zeros(shape=(n,n))
        
        for edgeA, edgeB in all_combinations:

            probability_experiment=random.random()

            if (probability_experiment<p):

                matrix[edgeA][edgeB]=1
                matrix[edgeB][edgeA]=1
    
        return matrix
```
